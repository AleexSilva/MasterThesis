```python
# Ratio de unicidad
def uniqueRate(data,initialColumn):    
  d = pd.DataFrame(columns=["Cantidad_√∫nicos","Ratio_√∫nicos"],index=data.columns)
  x = data.columns

  for i in data.columns:
    d.loc[i,"Cantidad_√∫nicos"]=data[i].nunique()

  d["Ratio_√∫nicos"]= d["Cantidad_√∫nicos"] / data.shape[0]
  d["Ratio_√∫nicos"] = d["Ratio_√∫nicos"].astype(float) *100
  d["Ratio_√∫nicos"] = round(d["Ratio_√∫nicos"],2).astype(str) + "%"
  return d.sort_values(by="Cantidad_√∫nicos",ascending=False)


uniqueRate(df,0)
```

## Limpieza de  datos

```python
# Cambios en la columna "Fecha".

df['Fecha'] = df['Fecha'].astype(str)
df['Hora']  = df.Fecha.str[10:13] + "hs" #Se crea una columna nueva para saber la hora en la que se cre√≥ el tweet
df['Mes']  = df.Fecha.str[5:7] #Se crea una columna nueva para saber la hora en la que se cre√≥ el tweet
df['Fecha']  = df.Fecha.str[0:10] #Se crea una columna para extraer s√≥lo la fecha de la craci√≥n del tweet
df['Fecha']= df['Fecha'].apply(pd.to_datetime)

di={'01':'Enero','03':'Marzo','02':'Febrero','04':'Abril','05':'Mayo','06':'Junio'}
df.replace({'Mes':di},inplace=True)
df.head()
```

```python
df.shape
```

```python
df['Descripcion_autor']=df.Descripcion_autor.str.upper()

words=['REVISTA','DIARIO','NOTICIA','NEWS']

for word in words:
    df=df.drop(df[(df['Descripcion_autor'].str.contains(word))].index)


words2=['¬°Firm√° la petici√≥n!','Infobae']

for word2 in words2:
    df=df.drop(df[(df['Texto'].str.contains(word2))].index)


words3=['dlvr.it','Foursquare Swarm','Foursquare']

for word3 in words3:
    df=df.drop(df[(df['Fuente']==word3)].index)
```

## Limpieza de datos - Texto

```python
# Se crea una funci√≥n para limpiar el texto presente en los tweets
def LimpiarTexto(texto):
    texto = re.sub('@[A-Za-z0‚Äì9]+', '', texto) # Eliminar "#" del texto
    texto = re.sub('#', '', texto) # Eliminar hashtags (#)
    texto = re.sub('@[\S]*', '', texto) #Eliminar menciones (@)
    texto = re.sub('\\n', '', texto) # Eliminar el caracter '\n'
    texto = re.sub('https?:\/\/\S+', '', texto) # Eliminar Hiperlinks (https:\)
    return texto

# Limpieza de tweets
df['Texto'] = df['Texto'].apply(LimpiarTexto)
df['Texto'] = df['Texto'].str.strip()
df['Texto'] = df['Texto'].str.strip('*')
```

```python
df = df.reset_index()
#Nueva Variable: Cantidad de caracteres en un Tweet

longitud=[]
for i in range(0,len(df.Texto)):
    longitud.append(len(df.Texto[i]))

df['Long_Texto'] = longitud
```
